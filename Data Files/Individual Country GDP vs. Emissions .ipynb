```python
#Dependecies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
fig, ax1 = plt.subplots()
```

```python
#Import CSV file
emission = pd.read_csv("emissions_data_clean.csv", encoding="ISO-8859-1", index_col=0)
gdp_df = pd.read_csv("Global_US_GDP.csv", skiprows = 4, skipinitialspace=True)
```

```python
#drop irrelevant columns names
clean_gdp = gdp_df.drop(['Indicator Name','Indicator Code', 'Unnamed: 65'], axis = 1)

#drop all rows up until 1990 to align with other data emissions data
for x in range(1960,1990):
    clean_gdp = clean_gdp.drop([f'{x}'],axis = 1)

#remove rows with any NaN values
clean_gdp = clean_gdp.dropna(how = 'any')
clean_gdp

#transpose columns
clean_gdp_t = clean_gdp.T
clean_gdp_t.columns = (clean_gdp_t.iloc[0])
clean_gdp_t = clean_gdp_t.drop(clean_gdp_t.index[0])
clean_gdp_t = clean_gdp_t.drop(clean_gdp_t.index[0])

clean_gdp_t.rename(columns={"Country Name" :"Year"})
clean_gdp_t
```

```python
#emission_clean = emission.drop(['Region','Income Gro



emission_clean = emission.T
emission_clean.rename(columns={"Country/Region" :"Year"})
emission_clean.columns = (emission_clean.iloc[0])
emission_clean = emission_clean.drop(emission_clean.index[0])


emission_clean
#emission_clean = emission_clean.drop(['Palau'], axis = 1)
#emission_clean = emission_clean.astype(float, errors = 'raise')

#flse values in data prevented converting the whole thing to floats...
emission_clean['China'] = emission_clean["China"].astype(float, errors = 'raise')
emission_clean['Japan'] = emission_clean["Japan"].astype(float, errors = 'raise')
emission_clean['Australia'] = emission_clean["Australia"].astype(float, errors = 'raise')
emission_clean['United States'] = emission_clean["United States"].astype(float, errors = 'raise')
emission_clean['India'] = emission_clean["India"].astype(float, errors = 'raise')

emission_clean
```

```python
merged = pd.merge(emission_clean, clean_gdp_t, left_index=True, right_index=True)


merged
```

```python
plt.figure(figsize=(20,10))
plt.plot(clean_gdp_t["China"], label= "China")
plt.plot(clean_gdp_t["Japan"], label= "Japan")
plt.plot(clean_gdp_t["Australia"], label= "Australia")
plt.plot(clean_gdp_t["United States"], label= "United States")
plt.plot(clean_gdp_t["India"], label= "India")


plt.title("5x Country GDP Over Time")
plt.legend()
plt.xticks(rotation=90)
plt.tick_params(axis='x', which='major', labelsize=15)
plt.tick_params(axis='y', which='major', labelsize=15)
plt.xlabel("Year" , fontsize=15)
plt.ylabel("GDP (trillions)", fontsize=15)
gdp_plot = plt.show()
plt.savefig('5 Country GDP Snapshot.png')
```
