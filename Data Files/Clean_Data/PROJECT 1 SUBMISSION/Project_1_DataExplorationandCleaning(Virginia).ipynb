### Reading Regions csv file and opening to understand the content

```python
# Dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Import CSV file
## I was getting a UnicodeDecodeError:'utf-8', and used 'encoding="ISO-8859-1' to solve the error
Region_income = pd.read_csv("Region_Income.csv",encoding="ISO-8859-1")
```

```python
Region_income.head()
```

```python
Region_income
```

```python
#Rename column
Region_income.rename(columns={'TableName': 'CountryName'}, inplace=True)
```

```python
#check
Region_income.head()
```

```python
#Drop a column
#I dropped the column 'special notes' and 'Country Code' since I dont think we need the information 
#Region_income.drop('SpecialNotes', axis =1, inplace =True)
columns= ['SpecialNotes', 'Country Code']
Region_income.drop(columns, inplace =True, axis =1)
```

```python
#check
Region_income.head()
```

```python
##Remove countries with NAN  
Region_income.dropna(inplace = True)
```

```python
##Change the order of the data frame
Region_income = Region_income.reindex(['CountryName', 'Region','IncomeGroup'], axis =1)
```

```python
Region_income.head()
```

```python
Region_income.rename(columns = {'CountryName':'Country'}, inplace = True)
```

```python
Region_income.head()
```

```python
Region_income.to_csv('Cleaner_Region_Income_data.csv')
```

## Reading GDP csv file cleaned by another team member

```python
GDP_df = pd.read_csv("clean_gdp.csv",encoding="ISO-8859-1")
```

## Merge the two dataframes (Region and GDP DF) using an inner join

```python
## Merge two dataframes (Region and GDP DF) using an inner join
merged_df = pd.merge(Region_df, GDP_df, on="Country")
merged_df.head()
```

```python
#drop irrelevant columns names
merged_df = merged_df.drop(['Country Code', 'Unnamed: 0_x', 'Unnamed: 0_y'], axis = 1)
```

```python
merged_df.head()
```

```python
merged_df.to_csv('Merged_GDP-REGION.csv', index=False, sep=',')
```
