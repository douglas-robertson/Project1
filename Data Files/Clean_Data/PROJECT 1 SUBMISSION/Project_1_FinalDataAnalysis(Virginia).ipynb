```python
# Dependencies
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Emission per region over time analysis

```python

df = pd.read_csv("Merged_emissions_region.csv")
df.head()
```

```python

#Create a heat map
piv = pd.pivot_table(df,index=["Region"])
#plot pivot table as heatmap using seaborn
ax = sns.heatmap(piv,cmap="gnuplot",fmt='.2f', linewidths=1, linecolor='white')
ax.collections[0].colorbar.set_label("MTCo2e")
plt.setp( ax.xaxis.get_majorticklabels(), rotation=90 )
ax.set_xlabel("Year", size=12)
ax.set_ylabel("Region",size=12)
plt.title(" Regions Emmisions over time", size=12)
plt.tight_layout()
sns.set(rc = {'figure.figsize':(20,10)})
plt.savefig('Regions Co2 Emmision over time.jpg',dpi=150,)
```

```python
sns.heatmap(piv.corr(), annot=True, fmt=".2f")
plt.setp( ax.xaxis.get_majorticklabels(), rotation=90 )
plt.title(" Correlation emmission over time", size=12)
ax.set_xlabel("Year", size=12)
ax.set_ylabel("year",size=12)
sns.set(rc = {'figure.figsize':(20,8)})
plt.savefig('correlation emmission over time.jpg',dpi=150,)
```

## GDP per region over time analysis

```python
#Read the csv file
df = pd.read_csv("Merged_GDP-REGION.csv")
df.head()
```

```python
#Create a heat map
piv = pd.pivot_table(df,index=["Region"])
#plot pivot table as heatmap using seaborn
ax = sns.heatmap(piv,cmap="gnuplot",fmt='.2f', linewidths=1, linecolor='white')
ax.collections[0].colorbar.set_label("GDP(Trillions)")
plt.setp( ax.xaxis.get_majorticklabels(), rotation=90 )
ax.set_xlabel("Year", size=12)
ax.set_ylabel("Region",size=12)
plt.title(" Regions GDP over time", size=12)
plt.tight_layout()
sns.set(rc = {'figure.figsize':(20,10)})
plt.savefig('Regions GDP-over time.jpg',dpi=150,)
```

```python
#Create a correlation heatmap
sns.heatmap(piv.corr(), annot=True, fmt=".2f")
plt.setp( ax.xaxis.get_majorticklabels(), rotation=90 )
plt.title(" Correlation GDP over time", size=12)
ax.set_xlabel("Year", size=12)
ax.set_ylabel("year",size=12)
sns.set(rc = {'figure.figsize':(20,8)})
plt.savefig('correlation GDP over time.jpg',dpi=150,)
```
