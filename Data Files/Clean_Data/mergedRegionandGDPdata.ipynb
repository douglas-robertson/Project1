```python
# Dependencies
import pandas as pd
```

```python
Region_df = pd.read_csv("Cleaner_Region_income_data.csv",encoding="ISO-8859-1")
```

```python
Region_df.head()
```

```python
GDP_df = pd.read_csv("clean_gdp.csv",encoding="ISO-8859-1")
```

```python
GDP_df.head()
```

```python
# Merge two dataframes using an inner join
merged_df = pd.merge(Region_df, GDP_df, on="Country")
merged_df.head()
```

```python
merged_df.to_csv('merged_GDPandREGION.csv')
```
